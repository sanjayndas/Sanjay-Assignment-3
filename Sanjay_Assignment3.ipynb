```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib as plt
import statistics as stat
```

**Average Scores based on Gender and Test Preparation Course**

```python
mmath_avg = data.loc[data['gender']=='male','math score'].mean()
fmath_avg = data.loc[data['gender']=='female','math score'].mean()

mread_avg = data.loc[data['gender']=='male','reading score'].mean()
fread_avg = data.loc[data['gender']=='female','reading score'].mean()

mwrite_avg = data.loc[data['gender']=='male','writing score'].mean()
fwrite_avg = data.loc[data['gender']=='female','writing score'].mean()

print('Mean marks for male and female :')
gmean_df = pd.DataFrame({'Math': [mmath_avg, fmath_avg],
                        'Reading': [mread_avg, fread_avg],
                         'Writing': [mwrite_avg, fwrite_avg]},
                       index=['Male','Female'])
gmean_df
```

```python
nmath_avg = data.loc[data['test preparation course']=='none','math score'].mean()
cmath_avg = data.loc[data['test preparation course']=='completed','math score'].mean()

nread_avg = data.loc[data['test preparation course']=='none','reading score'].mean()
cread_avg = data.loc[data['test preparation course']=='completed','reading score'].mean()

nwrite_avg = data.loc[data['test preparation course']=='none','writing score'].mean()
cwrite_avg = data.loc[data['test preparation course']=='completed','writing score'].mean()

print('Mean marks for Test Preparation Course :')
tmean_df = pd.DataFrame({'Math': [nmath_avg, cmath_avg],
                        'Reading': [nread_avg, cread_avg],
                         'Writing': [nwrite_avg, cwrite_avg]},
                       index=['None','Completed'])
tmean_df
```

**Top 25% based on Math Score**

```python

top = np.percentile(data['math score'], 75)
print('Bonus points can be given to students with Math Score above ', top)
topdf = data[data['math score']>=top]
print('List of students in top 25% based on math score:')
topdf
```
